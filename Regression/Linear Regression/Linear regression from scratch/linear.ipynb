```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.loadtxt("data.txt", delimiter = ",")
print(data.shape)
xx=data[:,0].reshape(45,1)
yy=data[:,1].reshape(45,1)
x_train=data[:30,:1]
x_train.shape
```

```python
y_train=data[:30,-1].reshape(30,1)
print(y_train.shape)
y_train
```

```python
x_test=data[-15:,:1]
x_test
y_test=data[-15:,-1].reshape(15,1)
y_test
```

```python
plt.plot(x_train,y_train,'o')
```

```python

def model(X, Y, learning_rate, iteration):
    costs=[]
    theta=np.zeros((1,1))
    m=Y.size
    for i in range(iteration):
        y_pred=np.dot(X,theta)
        cost=(1/2*m)*np.sum((np.square(y_pred-Y)))
        d_theta=(1/m)*np.dot(X.T,y_pred-Y)
        theta=theta-(learning_rate*d_theta)
        costs.append(cost)
    return theta,costs
```

```python
iteration = 100
learning_rate = 0.00000005
theta, cost_list = model(x_train, y_train, learning_rate = learning_rate,
iteration = iteration)
```

```python
# Y=((m*x_train)+c)
# plt.plot(x_train,Y,'r')
# plt.plot(x_train,y_train,'o')
# plt.show()
plt.title("Cost function vs Iteration")
plt.plot(np.array(cost_list),np.arange(0, iteration))
plt.show()
plt.title("x_train VS y_train")
plt.plot(x_train,y_train,'o')
plt.plot(x_train,np.dot(x_train,theta),'r')
plt.show()
plt.title("x_test VS y_test")
plt.plot(x_test,y_test,'o')
plt.plot(x_test,np.dot(x_test,theta),'r')
plt.show()
plt.title("x vs y [Overall]")
plt.plot(xx,yy,'o')
plt.plot(xx,np.dot(xx,theta),'r')
```
